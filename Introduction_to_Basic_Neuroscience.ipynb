## Electrochemical Potential

Neurons communicate <b>electrochemically </b> i.e. chemicals causes electrical signal. The chemical elements in the body are found in the form of ions and is no different in the nervous system. The important ions are sodium and pottasium (both positively charged +), Calcium (double positive ++) and Chlorine (- negatively charged) ions. Please note there is also a significant negatively charged proteins inside the neuron. The concentration of these ions inside and outside of the nerve cell varies and are maintaned by a membrane that is selectively permeable for certain ions.


## Resting Membrane Potential
<p> When a neuron is at rest, the inside of the neuon is negatively charged relative to the outside (extracellular space or matrix). This balance is maintained by specialized channels that are selective for specific ions. We will concentrate on 2 ions, Na+ and K+, There are specific ion channels for these, the Sodium channel and the Pottasium channel.</p>
<p> In addition to these channels, there is a pump that maintains and seperates the ions. In general, the majority of sodium ions is maintained outside the cell and the pottasium is maintianed inside the cell. </p>
<p> This results in a voltage difference of ~-70mV. We call this the resting membrane potential

## Action potential
An action potential is how a neuron conveys information. The information is generally one that moves away from the cell body or the soma through the axon. Neuroscientists often use the word "spike" or "impulse" to describe the action potential. Put simply, action potential occurs when a stimulus causes the membrane potential to move from resting potential (-70 mV) towards zero or <b> depolarizes</b>. This process is not linear, when the membrane potential reaches about -55mV (<b>threshold</b>), there is a rapid movement towards zero and beyond positive, this is called <b> the action potential </b>.
<p> It is important to note that once the threshold is reached, there is no small or big action potential. It is all or none. If the threshold is not reached an action potential cannot be reached.

##Mechanism of action potential generation
Action potentials are generated as a direct result of movement of different ions cross the neuron membrane. 
<ul> -  A stimulus (from a receptor or from other neurons) causes sodium channels to open. Because there are many more sodium ions on the outside, and the inside of the neuron is negative relative to the outside, sodium ions rush into the neuron. Remember, sodium carries a positive charge, so the neuron becomes more positive and becomes depolarized. </ul>

<ul> - Pottasium channel takes longer to open. When they do open, potassium rushes out of the cell, reversing the depolarization. Also at about this time, sodium channels start to close. This causes the action potential to go back toward -70 mV (a repolarization). The action potential actually shoots past -70 mV (a hyperpolarization) because the potassium channels stay open a bit too long. Gradually, the sodium-pottasium pump ensures that the ion concentrations go back to resting levels and the cell returns to -70 mV.
